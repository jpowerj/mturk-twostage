```python
def get_hit_for_worker(worker_id):
    for cur_hit in all_hits:
        hit_title = cur_hit['Title']
        hit_id = cur_hit['HITId']
        hit_creation = cur_hit['CreationTime']
        print(f"Checking {hit_title}: {hit_id} (Created {hit_creation})")
        # Get the list of worker_ids who submitted this hit
        hit_submissions = mtglobals.get_hit_submissions(client, cur_hit['HITId'])
        sub_worker_ids = [s['WorkerId'] for s in hit_submissions]
        print(sub_worker_ids)
        # Check if the worker_id we want is in it
        if worker_id in sub_worker_ids:
            return cur_hit, hit_submissions
```
